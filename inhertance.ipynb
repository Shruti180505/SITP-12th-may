 ### Inheritance in Python is a feature of Object-Oriented Programming (OOP) that allows one class to reuse the properties and methods of another class.

* The existing class is called the parent class (or base class).
* The new class is called the child class (or derived class).

      Inheritance helps in:

1. Code Reusability → reuse existing code from parent class
2. Reducing Code Duplication → no need to write same methods again
3. Easy Maintenance → changes in parent class affect child classes automatically
4. Extending Functionality → child class can add new features
5. Method Overriding → child class can modify parent behavior
6. Better Organization of Code → creates logical class hierarchy
7. Achieving Polymorphism → same method behaves differently in different classes

```python
# Multiple inheritance
class Developer:
    def developer_info(self):
        print("I am Developer")
class Manager:
    def manager_info(self):
        print("I am Manager")
class TeamLeader(Developer,Manager):
    def teamleader_info(self):
        print("I am Team Leader")

tl=TeamLeader()
tl.teamleader_info()
tl.manager_info()
tl.developer_info()
```

```python
class Company1:
    def company1_data(self):
        print("the sales of the company has increased by 60%")
class Company2:
    def company2_data(self):
        print("the sales of the company has increased by 70%")
class Company3(Company1, Company2):
    def company3_data(self):
        print("the sales of the company has not been increased ")
c3=Company3()
c3.company1_data()
c3.company2_data()
c3.company3_data()
```
