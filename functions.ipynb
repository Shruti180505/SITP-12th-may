```python
def find(lst1):
    for x in lst1:

        if x%2==0:
            print("the number is even",x)
        elif x<0:
            print("the number is negative",x)
        else:
            print("the number is odd",x)
        

lst1=[1,2,3,4,5,-1]
find(lst1)


```

```python
def grade(g):
    if g>=90 and g<=100:
        print("A++")
    elif g>=80 and g<90:
        print("A")
    elif g>=70 and g<80:
        print("B++")
    elif g>=60 and g<70:
        print("B")
    elif g>=50 and g<60:
        print("C++")
    elif g>=40 and g<50:
        print("D++")
    elif g<40:
        print("fail")
    else:
        print("enter a valid marks")


g=int(input("enter the marks of student"))
grade(g)
```

```python
def prime():
    lst1=[]
    lst1.append(1)
    for x in range(2,100):
        count=0

        for y in range(1,x+1):
            if x%y==0:
                count+=1
           
        if count==2:
           lst1.append(x)

    
    print(lst1)

prime()

```
