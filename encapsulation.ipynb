```python
class Student:
  def __init__(self, name, classes, rollno, branch):
   self.name=name
   self.classes=classes
   self.rollno=rollno
   self.branch=branch

  def show(self):
     print("the name of student is", self.name)
     print("the class of student is", self.classes)
     print("the rollno of student is", self.rollno)
     print("the branch of student is", self.branch)

s=Student("shruti", "cse","23eemcs115","btech")
s.show()
       
```

```python
class Hod:
    def __init__(self,hod_info):
      self.hod_info=hod_info
    
class Teacher(Hod):
   def __init__(self,hod_info,teacher_info):
      super(). __init__(hod_info)

      self.teacher_info=teacher_info

class Student(Teacher):
    def __init__(self, hod_info,teacher_info,student_info):
       super() .__init__(hod_info,teacher_info)

       self.student_info=student_info

    def show(self):
          print("the age of Hod is", self.hod_info)
          print("the age of teacher is", self.teacher_info)
          print("the age of student is", self.student_info)

obj=Student("42", "35","12")
obj.show()

          
```
