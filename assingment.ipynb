```python
class FactorialCalculator:
    factorial=1
    def calculate(self):
     num=int(input("enter the factorial"))
     for x in range(num,1,-1):
         self.factorial*=x

     print("The factorial of the number is", self.factorial)
 
f=FactorialCalculator()
f.calculate()
```

```python
class PrimeNumberChecker:
    def check(self):
        num=int(input("enter the number"))
        count=0
        for x in range(1,num+1):
            if num%x==0:
                count+=1
        if count==2:
          print("the number is prime")
        else:
          print("the number is not prime")

p=PrimeNumberChecker()
p.check()
    
```

```python
class Person:
    def __init__(self,name,age):
     self.name=name
     self.age=age

    def show(self):
        print("the name of the student is", self.name)
        print("the age of the student is", self.age)

class Teacher(Person):
   def __init__(self,name,age,branch,rollno):
      super().__init__(name,age)

      self.branch=branch
      self.rollno=rollno

   def display(self):
      print("the name of the student is", self.name)
      print("the age of the student is", self.age)
      print("the branch of the student is", self.branch)
      print("the rollno of the student is", self.rollno)

obj=Teacher("shruti",22,"CSE","23EEMCS115")
obj.show()
obj.display()
```

```python
class Vechile:
    def __init__(self,vechile):
        self.vechile=vechile

class SportsCar(Vechile):
    def __init__(self,vechile,model):
        super(). __init__(vechile)
        self.model=model

class Car(SportsCar):
    def __init__(self,vechile,model,color):
        super(). __init__(vechile,model)
        self.color=color
    
    def show(self):
        print("the vechile is a", self.vechile)
        print("the model of the car is", self.model)
        print("the color of the car is", self.color)

obj=Car("4-wheeler", "F8", "red")
obj.show()
```

```python
class Calculator:
    def __init__(self,a,b,op):
        self.a=a
        self.b=b
        self.op=op
    
    def calculate(self):

     match self.op:
      case 1: 
       print("the addition of two numbers is", self.a + self.b)
      case 2:
    
        print("the subtraction of two numbers is", self.a - self.b)
      case 3:
    
        print("the multiplication of two numbers is", self.a * self.b)
      case 4:
   
        print("the divison of two numbers is", self.a % self.b)

a=int(input("enter the first number"))
b=int(input("enter the second number"))
print("1=addition")
print("2=subtraction")
print("3=multiplication")
print("4= divison")
op=int(input("enter the number of the operation you want to perform"))
obj=Calculator(a,b,op)
obj.calculate()


        
```

```python
class BankAccount():
    currentbalance=5000

    def __init__(self,balance,op):
        self.balance=balance
        self.op=op
    def bank(self):
      
        match self.op:
         case 1:
              print("the money has been deposited to your account")
              BankAccount.currentbalance+=self.balance
         case 2:
                print("the withdrawal has been initiated")
                BankAccount.currentbalance-=self.balance
         case 3:
                print("the current available baalnce is", BankAccount.currentbalance)

         case 4:
                  print("Thank you")
                  return False
         case _:
                  print("invalid operation")

        return True

while True:   


 print("\n------BANK MENU-----")
 print("1=deposit")
 print("2=withdrawal")
 print("3=check balance")
 print ("4=exit")
 op=int(input("enter the operation you want to perform"))
 if op==1:
     deposit=int(input("enter the deposit amount"))
     obj =BankAccount(deposit,op)
 elif op==2:
     withdrawal=int(input("enter the withdrawal amount"))
     obj=BankAccount( withdrawal,op)
 else:
     obj=BankAccount(0,op)

 status=obj.bank()
 if status == False:
   break
```
