# NumPy Summary Table

| Topic | Description | Example |
|---|---|---|
| NumPy | Numerical Python library | `import numpy as np` |
| Array | Stores multiple values | `np.array([1,2,3])` |
| 1D Array | Single dimensional array | `[1 2 3]` |
| 2D Array | Rows and columns | `[[1,2],[3,4]]` |
| ndarray | Main NumPy object | `type(arr)` |
| ndim | Checks dimensions | `arr.ndim` |
| shape | Returns rows & columns | `arr.shape` |
| size | Total elements count | `arr.size` |
| dtype | Data type of array | `arr.dtype` |
| zeros() | Creates zero array | `np.zeros((2,2))` |
| ones() | Creates ones array | `np.ones((2,2))` |
| arange() | Creates range array | `np.arange(1,10)` |
| linspace() | Equal spaced values | `np.linspace(1,10,5)` |
| reshape() | Changes array shape | `arr.reshape(2,3)` |
| Indexing | Access single element | `arr[0]` |
| Slicing | Access multiple elements | `arr[1:4]` |
| Addition | Adds arrays | `a+b` |
| Subtraction | Subtract arrays | `a-b` |
| Multiplication | Multiply arrays | `a*b` |
| Division | Divide arrays | `a/b` |
| sum() | Finds total sum | `np.sum(arr)` |
| max() | Finds maximum value | `np.max(arr)` |
| min() | Finds minimum value | `np.min(arr)` |
| mean() | Finds average | `np.mean(arr)` |
| Advantages | Fast & memory efficient | Used in Data Science |
| Applications | ML, AI, Statistics | Data Analysis |
