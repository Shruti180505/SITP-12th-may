| Function           | Work                          | Example                       |
| ------------------ | ----------------------------- | ----------------------------- |
| os.getcwd()      | Get current working directory | os.getcwd()                 |
| os.chdir(path)   | Change current directory      | os.chdir("D:/Python")       |
| os.listdir()     | Show all files and folders    | os.listdir()                |
| os.mkdir()       | Create one folder             | os.mkdir("Student")         |
| os.makedirs()    | Create multiple folders       | os.makedirs("Data/Student") |
| os.rmdir()       | Delete empty folder           | os.rmdir("Student")         |
| os.removedirs()  | Delete nested empty folders   | os.removedirs("A/B/C")      |
| os.remove()      | Delete file                   | os.remove("data.txt")       |
| os.rename()      | Rename file/folder            | os.rename("old","new")      |
| os.path.exists() | Check file/folder exists      | os.path.exists("Data")      |
| os.path.isfile() | Check file exists             | os.path.isfile("a.txt")     |
| os.path.isdir()  | Check folder exists           | os.path.isdir("Student")    |
| os.path.join()   | Join file paths               | os.path.join("A","B")       |
